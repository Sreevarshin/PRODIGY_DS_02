```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic_df = pd.read_csv(r'C:\Users\HP\Downloads\train (1).csv')
```

# Step 1: Data Cleaning

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
# Fill missing Age values with the median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python
# Fill missing Embarked values with the mode
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
```

```python
# Drop Cabin column as it has too many missing values
titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
# Verify missing values have been handled
print(titanic_df.isnull().sum())
```

```python
# Convert Categorical Variables to Numerical
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df['Embarked'] = titanic_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Feature Engineering (if necessary, e.g., create FamilySize from SibSp and Parch)
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
```

# Step 2: Exploratory Data Analysis (EDA)

```python
# Visualize Survived
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Counts')
plt.show()
```

```python
# Survival Rate by Gender
plt.figure(figsize=(6, 4))
sns.barplot(data=titanic_df, x='Sex', y='Survived')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival Rate by Passenger Class
plt.figure(figsize=(6, 4))
sns.barplot(data=titanic_df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Survival Rate by Age
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Age', hue='Survived', bins=30, kde=True)
plt.title('Survival Rate by Age')
plt.show()
```

```python
# Survival Rate by Embarked
plt.figure(figsize=(6, 4))
sns.barplot(data=titanic_df, x='Embarked', y='Survived')
plt.title('Survival Rate by Embarked')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(titanic_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
